## Kalman filter

```python
!pip install pykalman
```

```python
from pykalman import KalmanFilter
import numpy as np

# Khởi tạo bộ lọc Kalman
kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)

# Áp dụng bộ lọc Kalman cho từng cột dữ liệu
data_col[f'{col1}_kalman'] = kf.smooth(data_col[str(col1)].values)[0]
data_col[f'{col2}_kalman'] = kf.smooth(data_col[str(col2)].values)[0]
data_col[f'{col3}_kalman'] = kf.smooth(data_col[str(col3)].values)[0]

# Trực quan hóa kết quả sau khi áp dụng bộ lọc Kalman
fig, ax = plt.subplots(figsize = (10,5))
sns.lineplot(ax=ax, data=data_col, x="date", y=str(col1), label=col1, color ='b')
sns.lineplot(ax=ax, data=data_col, x="date", y=f'{col1}_kalman', label=f'{col1}_kalman', color ='r')
plt.title('Data with Kalman Filter')
plt.legend()
plt.grid(True)
plt.show()
```

## HMM model

```python
!pip install hmmlearn
```

```python
from hmmlearn import hmm
import numpy as np

# Định nghĩa mô hình HMM
model = hmm.GaussianHMM(n_components=2, covariance_type="diag")

# Huấn luyện mô hình với dữ liệu
series = data_col[f'{col1}_kalman']
X = np.array(series).reshape(-1, 1)
model.fit(X)

# Dự đoán các trạng thái ẩn
hidden_states = model.predict(X)

# Xuất kết quả
print("Trạng thái ẩn:", hidden_states)
print("Xác suất ban đầu:", model.startprob_)
print("Ma trận chuyển đổi:", model.transmat_)
print("Trung bình của các trạng thái:", model.means_)
print("Trung vị của các trạng thái:", model.covars_)
```

```python
# Trực quan hóa trạng thái ẩn
plt.figure(figsize=(10, 5))
sns.scatterplot(data=data_col, x='date', y=f'{col1}_kalman', hue=hidden_states, palette='tab10')
plt.title(f'Hidden States for {col1}_kalman')
plt.show()
```
